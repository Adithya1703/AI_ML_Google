| Normalization Technique | Formula                                                                 | When to Use                                                                                     |
|--------------------------|------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------|
| Linear Scaling           | $x^* = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$     | When data needs to be scaled to a fixed range (e.g., [0, 1]) and there are no extreme outliers. |
| Z-Score Scaling          | $Z = \frac{X - \mu}{\sigma}$                                           | When data needs to be standardized to have a mean of 0 and a standard deviation of 1.           |
| Log Scaling              | $x^* = \log(X + 1)$                                                   | When data has a wide range or is skewed, and you want to reduce the impact of large values.     |
| Clipping                 | $x^* = \min(\max(X, \text{lower}), \text{upper})$                     | When you want to limit the range of data to remove extreme outliers.                           |

```python
def find_outliers_iqr(df, column):
    """Finds outliers in a column using the IQR method and returns a DataFrame with a 'Reason' column."""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].copy()  # Create a copy to avoid SettingWithCopyWarning
    outliers['Reason'] = [f'Value outside IQR ({lower_bound:.2f}, {upper_bound:.2f})' for _ in range(len(outliers))]
    return outliers


def find_outliers_zscore(df, column, threshold=3):
    """Finds outliers in a column using the Z-score method and returns a DataFrame with a 'Reason' column."""
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = df[z_scores > threshold].copy()  # Create a copy to avoid SettingWithCopyWarning
    outliers['Reason'] = [f'Z-score > {threshold}' for _ in range(len(outliers))]
    return outliers

def find_outliers_percentile(df, column, lower_percentile=1, upper_percentile=99):
    """Finds outliers in a column using the percentile method and returns a DataFrame with a 'Reason' column."""
    lower_bound = df[column].quantile(lower_percentile / 100)
    upper_bound = df[column].quantile(upper_percentile / 100)
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].copy()  # Create a copy to avoid SettingWithCopyWarning
    outliers['Reason'] = [f'Value outside percentile range ({lower_bound:.2f}, {upper_bound:.2f})' for _ in range(len(outliers))]
    return outliers
```

```python
for column in training_df.select_dtypes(include=np.number).columns:
    print(f"Column: {column}")
    outliers_iqr = find_outliers_iqr(training_df, column)
    print(f"  IQR Method: {len(outliers_iqr)} outliers")
    outliers_zscore = find_outliers_zscore(training_df, column)
    print(f"  Z-Score Method: {len(outliers_zscore)} outliers")
    outliers_percentile = find_outliers_percentile(training_df, column)
    print(f"  Percentile Method: {len(outliers_percentile)} outliers")
```

```python
for column in training_df.select_dtypes(include=np.number).columns:
    print(f"Column: {column}")

    # IQR Method
    outliers_iqr = find_outliers_iqr(training_df, column)
    print(f"  IQR Method: {len(outliers_iqr)} outliers")
    if not outliers_iqr.empty:  # Check if there are any outliers
        print("    Outlier Rows:")
        print(outliers_iqr)  # Print the outlier rows
    else:
        print("    No outliers found.")

    # Z-Score Method
    outliers_zscore = find_outliers_zscore(training_df, column)
    print(f"  Z-Score Method: {len(outliers_zscore)} outliers")
    if not outliers_zscore.empty:  # Check if there are any outliers
        print("    Outlier Rows:")
        print(outliers_zscore)  # Print the outlier rows
    else:
        print("    No outliers found.")

    # Percentile Method
    outliers_percentile = find_outliers_percentile(training_df, column)
    print(f"  Percentile Method: {len(outliers_percentile)} outliers")
    if not outliers_percentile.empty:  # Check if there are any outliers
        print("    Outlier Rows:")
        print(outliers_percentile)  # Print the outlier rows
    else:
        print("    No outliers found.")
```
